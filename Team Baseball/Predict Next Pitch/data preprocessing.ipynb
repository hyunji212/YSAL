```python
cutoff=data[['pitch_name','game_date','batter','pitcher','player_name','description','events','zone','stand','p_throws','plate_x','plate_z','inning','pitch_number','home_team','away_team','inning_topbot']]
cutoff
```

```python
#p_throws change
cutoff.loc[cutoff['p_throws'] == 'L', 'p_throws'] = 1
cutoff.loc[cutoff['p_throws'] == 'R', 'p_throws'] = 2

#stand
cutoff.loc[cutoff['stand'] == 'L', 'stand'] = 1
cutoff.loc[cutoff['stand'] == 'R', 'stand'] = 2


#pitch_type
cutoff['pitch_name'].fillna('N', inplace=True) # nanì„ Nìœ¼ë¡œ ìˆ˜ì •
fastball = ['Cutter','4-Seam Fastball','Split-Finger' ] #Fastball
for keyword in fastball:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['pitch_name'].str.contains(keyword), 'pitch_name'] = '1'


curve= ['Knuckle Curve','Curveball', 'Slow Curve']#Curve
for keyword in curve:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['pitch_name'].str.contains(keyword), 'pitch_name'] = '2'


slider= ['Slider','Slurve','Sweeper']#Breaking ball
for keyword in slider:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['pitch_name'].str.contains(keyword), 'pitch_name'] = '3'

changeup= ['Changeup']#Breaking ball
for keyword in changeup:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['pitch_name'].str.contains(keyword), 'pitch_name'] = '4'

sinker= ['Sinker']#sinker
for keyword in sinker:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['pitch_name'].str.contains(keyword), 'pitch_name'] = '5'


etc= [ 'Pitch Out', 'N']#etc
for keyword in etc:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['pitch_name'].str.contains(keyword), 'pitch_name'] = '6'

#description
ball = ['ball', 'blocked_ball','hit_by_pitch'] #ball
for keyword in ball:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['description'].str.contains(keyword), 'description'] = 'B'

strike = ['called_strike', 'foul_tip','missed_bunt','swinging_strike','swinging_strike_blocked','foul','foul_bunt' ] #strike
for keyword in strike:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['description'].str.contains(keyword), 'description'] = 'S'

in_play = ['hit_into_play' ] #in-play
for keyword in in_play:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['description'].str.contains(keyword), 'description'] = 'X'

nothing = [ 'pitchout' ]
for keyword in ball:
    # í•´ë‹¹ í‚¤ì›Œë“œë¥¼ í¬í•¨í•˜ëŠ” í–‰ì„ ì„ íƒí•˜ì—¬ ê°’ì„ ë³€ê²½
    cutoff.loc[cutoff['description'].str.contains(keyword), 'description'] = 'P'

cutoff
```

```python
ground = ['field_out','grounded_into_double_play','double_play',
'field_error','force_out','sac_fly']
ground_out = (cutoff['description'] == 'X') & (cutoff['events'].isin(ground))
cutoff.loc[ground_out, 'description'] = 'G'

K = ['strikeout']
strike_out = (cutoff['description'] == 'S') & (cutoff['events'].isin(K))
cutoff.loc[strike_out, 'description'] = 'K'

hits = ['single','double','triple','home_run']
hit = (cutoff['description'] == 'X') & (cutoff['events'].isin(hits))
cutoff.loc[hit, 'description'] = 'H'

walks = ['hit_by_pitch','walk']
walk = (cutoff['description'] == 'B') & (cutoff['events'].isin(walks))
cutoff.loc[walk, 'description'] = 'W'

cutoff
```
