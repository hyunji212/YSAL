```python
from tensorflow.keras.losses import Huber
from tensorflow.keras.activations import elu,gelu,swish,tanh,softmax
from tensorflow.keras.optimizers import Adam,RMSprop
from keras.layers import Input, LSTM, Dense,GRU
import tensorflow as tf
from keras.layers import Flatten
from keras.layers import Dropout
from tensorflow.keras import optimizers
from keras import layers
from sklearn.model_selection import train_test_split
from keras import losses
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.models import Sequential

#    
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

# relu   
# sigmoid 0~1  
# softmax   

# #   Sigmoid , Tanh ,ReLU, Leaky ReLU, PReLU, ELU,  Maxout
elu = tf.nn.elu
# leaky_relu = tf.nn.leaky_relu # leaky_relu
# activation = elu # elu #leaky_relu#"relu"#"tanh" #"relu"=>0~1  
```

```python
model = keras.models.Sequential()
#  
# "swish"
model.add(layers.Dense(32, activation=gelu, input_dim=8, name='Hidden-1'))
#   Dropout(p) p       ,  
# model.add(layers.Dropout(0.2))
#  
model.add(layers.Dense(16, activation=gelu, name='Hidden-2'))
#   Dropout(p) p       ,  
# model.add(layers.Dropout(0.2))
# 
model.add(layers.Dense(1, activation='sigmoid'))
# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# model.compile(loss='categorical_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])


model.compile(optimizer=RMSprop(learning_rate = 0.001),loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
import os
from keras.callbacks import EarlyStopping, ModelCheckpoint
# earlystopping (patience )n epoch val_loss    .
early_stop = EarlyStopping(monitor='val_loss', patience=20)

model_path = 'model/'
filename = os.path.join(model_path,'DNN.h5')

checkpoint = ModelCheckpoint(filename, #filepath
                             monitor='val_loss',#     => val_loss loss    
                             verbose=1, #  1   
                             save_best_only=True, # True         
                             save_weights_only=True, # True      
#                              save_freq = BATCH_SIZE, # 'epoch'  ,    . integer  ,     .
                             mode='auto'# val_acc  ,    .   max .  val_loss  , loss     .   min . auto  ,   min, max   .
                            )
```

```python
history = model.fit(X_train, y_train,
                                    epochs=1000,
                                    batch_size=10,
#                                     validation_data=(x_validation_WINDOW, y_validation_WINDOW),
                                    validation_data=(X_test, y_test),
                                    callbacks=[early_stop, checkpoint]) #   ,, L1,L2,    
```

```python
# summarize history for loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
